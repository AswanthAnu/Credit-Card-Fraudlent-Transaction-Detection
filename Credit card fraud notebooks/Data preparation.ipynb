```python
# Merging test_df and customer_df
test_df = pd.merge(test_df,customer_df[['CUSTOMER_ID', 'x_customer_id', 'y_customer_id', 'mean_amount',
       'std_amount', 'mean_nb_tx_per_day', 'available_terminals',
       'nb_terminals']],on='CUSTOMER_ID', how='left')
```
