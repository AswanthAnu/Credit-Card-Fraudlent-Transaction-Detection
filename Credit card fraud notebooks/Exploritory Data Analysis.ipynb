### Bivariate analysis

```python
# subsetting the data into valid and fraud

fraud = df[df.is_fraud == 1]
valid = df[df.is_fraud == 0]
```

```python
# transaction distribution of both fraud and valid
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# plot of fraud 
sns.distplot(x=fraud['trans_amount'], bins=12, color='red', ax=axs[0])
axs[0].set_title('Distribution of transaction amount in fraud')

# plot of valid 
sns.distplot(x=valid['trans_amount'], bins=12, color='green', ax=axs[1])
axs[1].set_title('Distribution of transaction amount in valid')
```

```python
# Largest transaction amount in valid and fraud
print('Largest transaction amount in fraud:', fraud.trans_amount.max())
print('Largest transaction amount in valid:', valid.trans_amount.max())
```

```python
# Smallest transaction amount in valid and fraud
print('Smallest transaction amount in fraud:', fraud.trans_amount.min())
print('Smallest transaction amount in valid:', valid.trans_amount.min())
```

```python
# Median transaction amount in valid and fraud
print('Median transaction amount in fraud:', fraud.trans_amount.median())
print('Median transaction amount in valid:', valid.trans_amount.median())
```

```python
# Boxplot of both fraud and valid
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# plot of fraud 
sns.boxplot(x=fraud['trans_amount'], color='red', ax=axs[0])
axs[0].set_title('Boxplot of transaction amount in fraud')

# plot of valid 
sns.boxplot(x=valid['trans_amount'],color='green', ax=axs[1])
axs[1].set_title('Boxplot of transaction amount in valid')
```

#### Insight

- Both the distribution of fraud and valid transactions are almost similar and right skewed.
- There maybe a chance that the transaction amount greater than 220 are fraud transactions.
- The smallest transaction in the fraud is 0.05 and valid is 0. So can be consider 0 as null, because there is no transactions that held with 0 amount.
- For both the valid and fraud the median is around 45.
- There are outliers present above around 160.

```python
# mean_amount distribution of both fraud and valid
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# plot of fraud 
sns.distplot(x=fraud['mean_amount'], bins=12, color='red', ax=axs[0])
axs[0].set_title('Distribution of mean transaction amount in fraud')

# plot of valid 
sns.distplot(x=valid['mean_amount'], bins=12, color='green', ax=axs[1])
axs[1].set_title('Distribution of mean transaction amount in valid')
```

#### Insight

- Fraud transactions happening in every range of transactions.
- The distribution is similar to uniform

```python
# mean_num_trans_per_day distribution of both fraud and valid
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# plot of fraud 
sns.distplot(x=fraud['mean_num_trans_per_day'], bins=4, color='red', ax=axs[0])
axs[0].set_title('Mean transaction amount per day in fraud')

# plot of valid 
sns.distplot(x=valid['mean_num_trans_per_day'], bins=4, color='green', ax=axs[1])
axs[1].set_title('Mean transaction amount per day in valid')
```

```python
# Boxplot of mean number of transaction amount per day of both fraud and valid
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# plot of fraud 
sns.boxplot(x=fraud['mean_num_trans_per_day'], color='red', ax=axs[0])
axs[0].set_title('Boxplot of transaction amount in fraud')

# plot of valid 
sns.boxplot(x=valid['mean_num_trans_per_day'],color='green', ax=axs[1])
axs[1].set_title('Boxplot of transaction amount in valid')
```

#### Insight

- Fraud transaction increases as the mean number of transaction per day increases as similar to valid.
- Most of the customers do 2-3.5 mean number of transaction per day.

```python
# number of terminals for a user distribution of both fraud and valid
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# plot of fraud 
sns.histplot(x=fraud['num_terminals'], bins=14, color='red', ax=axs[0])
axs[0].set_title('Number of terminals of users in fraud')

# plot of valid 
sns.histplot(x=valid['num_terminals'], bins=14, color='green', ax=axs[1])
axs[1].set_title('Number of terminals of users in valid')
```

```python

fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# plot of fraud 
sns.scatterplot(x=fraud['terminal_x_id'], y=fraud['terminal_y_id'], color='red', ax=axs[0, 0])
axs[0, 0].set_title('Terminals where fraud transactions', color='Blue')

# plot of valid 
sns.scatterplot(x=valid['terminal_x_id'], y=valid['terminal_y_id'], color='green', ax=axs[0, 1])
axs[0, 1].set_title('Terminals where valid transactions', color='Blue')

# plot of fraud 
sns.scatterplot(x=fraud['customer_x_id'], y=fraud['customer_y_id'], color='red', ax=axs[1, 0])
axs[1, 0].set_title('customers did fraud transactions', color='Blue')

# plot of valid 
sns.scatterplot(x=valid['customer_x_id'], y=valid['customer_y_id'], color='green', ax=axs[1, 1])
axs[1, 1].set_title('customers did valid transactions', color='Blue')

plt.subplots_adjust(top=1.2)
plt.show()
```

```python
df.columns
```

```python
# converting datatype of trans_datetime from object to datetime
df.trans_datetime = pd.to_datetime(df.trans_datetime)
```

```python
# Extracting year, month, day, hour, day_name
df['trans_month'] = df['trans_datetime'].dt.month
df['trans_day'] = df['trans_datetime'].dt.day
df['trans_hour'] = df['trans_datetime'].dt.hour
df['trans_dayname'] = df['trans_datetime'].dt.day_name()
```

```python
df.sample(5)
```

```python
fraud = df[df.is_fraud == 1]
valid = df[df.is_fraud == 0]
```

```python
# Transaction distribution according to month
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# plot of fraud 
sns.countplot(x=fraud['trans_month'],color='red', ax=axs[0])
axs[0].set_title('Count of fraud transaction in different months', color='Blue')

# plot of valid 
sns.countplot(x=valid['trans_month'],color='green', ax=axs[1])
axs[1].set_title('Count of valid transaction in different months', color='Blue')

plt.show()
```

```python
# Transaction distribution according to day
fig, axs = plt.subplots(1, 2, figsize=(16, 3))

# plot of fraud 
sns.countplot(x=fraud['trans_day'],color='red', ax=axs[0])
axs[0].set_title('Count of fraud transaction in different day', color='Blue')

# plot of valid 
sns.countplot(x=valid['trans_day'],color='green', ax=axs[1])
axs[1].set_title('Count of valid transaction in different day', color='Blue')

plt.show()
```

```python
# Transaction distribution according to dayname
fig, axs = plt.subplots(1, 2, figsize=(15, 4))

# plot of fraud 
sns.countplot(x=fraud['trans_dayname'],color='red', ax=axs[0])
axs[0].set_title('Count of fraud transaction in different day in a week', color='Blue')

# plot of valid 
sns.countplot(x=valid['trans_dayname'],color='green', ax=axs[1])
axs[1].set_title('Count of valid transaction in different day in a week', color='Blue')

plt.show()
```

```python
# Transaction distribution according to dayname
fig, axs = plt.subplots(1, 2, figsize=(15, 4))

# plot of fraud 
sns.countplot(x=fraud['trans_hour'],color='red', ax=axs[0])
axs[0].set_title('Count of fraud transaction in different hours', color='Blue')

# plot of valid 
sns.countplot(x=valid['trans_hour'],color='green', ax=axs[1])
axs[1].set_title('Count of valid transaction in different hours', color='Blue')

plt.show()
```

#### Insight

- The transaction during each month is uniform distribution for valid transactions and almost unifrom for fraud except there is a dip in November month.
- Almost 60K valid transaction take place at the every month and at the same time around 1.2K to 1.4K fraud transaction take place.
- The distribution of transaction per day also uniform. 
- The valid transactions and fraud transactions are also uniform in each day. With some small increase of fraud transaction in Tuesday to Friday.
- As normal, the transaction is normally distributed in hours per day. As the most transaction occurs high on the middle of the day for both valid and fraud.
